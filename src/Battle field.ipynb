```python
Mu = np.linspace(0,1,300)
na = np.array([2,1])
nA = np.array([2,3])

Prob = []

t0 = time.time()
for m in Mu:
    M = np.array([[1,-m],[-m,1]])
    R_mu = R(M)
    Prob.append(np.abs(aA(na,nA,R_mu))**2)
t1 = time.time()

fig, ax = plt.subplots(1,2,figsize=(15,4))
ax[1].plot(Mu,Prob)
ax[1].set_yscale('log')
ax[1].set_xscale('log')
ax[1].set_ylabel('Logarithm')

ax[0].plot(Mu,Prob)
ax[0].set_ylabel('Decimal')

ax[0].set_title('na = {}, nA = {}'.format(na,nA))
ax[1].set_title('time = {}'.format(t1-t0))

plt.show()
```
